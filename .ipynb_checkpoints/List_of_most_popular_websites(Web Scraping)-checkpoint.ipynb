##  - Importing the libraries

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import pandas as pd

print("import successful!!")
```

## - Fetching the data using GET request

```python
url = "https://en.wikipedia.org/wiki/List_of_most_popular_websites"
urlData = urlopen(url)
```

```python
url_html = urlData.read()
urlData.close()
```

```python
print(url_html)
```

## - Parsing the html using Beautiful soup 

```python
#beautiul soup parses the data given to it.
dataSoup = soup(url_html,'html.parser')
print(dataSoup)
```

```python
dataSoup.findAll('h1',{})
```

## - findAll('&lt;tag_name&gt;','&lt;{attribute}&gt;') method return you a list of all the searched tags present in the html data file

```python

usefulData = dataSoup.findAll('table',{'class':'wikitable'})[0]
print(len(usefulData))
```

```python
print(usefulData)
print(type(usefulData))
```

```python
l = 'Alexa top 50 global sites(As of April\xa016, 2020[update]) [3]'.split('[')
print(l[0]+')')
```

## - fetching useful information

```python
headings = usefulData.findAll('th')
columns = []

for data in headings:
    columns.append(data.text[:-1])

for indx,data in enumerate(columns):
    l = []
    l = columns[indx].split('(')
    columns[indx] = l[0]
```

```python
print(columns)
```

```python
rows = usefulData.findAll('tr')[1:]
print(len(rows))
```

```python
#creating rowData[] to store the row data of the table
rowData = []
for row in rows:
    currRowData = []
    rowdata = row.findAll('td')
    
    for indx,data in enumerate(rowdata):
        currRowData.append(data.text[:-1])
    
    rowData.append(currRowData)
```

## - Cleaning the data

```python
#data cleaning
for indx,row in enumerate(rowData):
    l = []
    k = []
    m = []
    if(rowData[indx][3] == '[data unknown/missing]'):
        rowData[indx][3] = 'N/A'
    
    l = rowData[indx][2].split('(')
    rowData[indx][2] = l[0]
    
    if(rowData[indx][3] == 'N/A[note 1]'):
        rowData[indx][3] = 'N/A'
    else:
        k = rowData[indx][3].split('(')
        rowData[indx][3] = k[0]
        
    if(rowData[indx][4]=='[data unknown/missing]'):
        rowData[indx][4] = 'N/A'
        
    m = rowData[indx][3].split(',')
    str = ""
    for val in m:
        str += val
    rowData[indx][3] = str

for row in rowData:
    print(row)
    
print(len(rowData))
```

## - Working with the CSV file *[writing the data into the csv file.]*

```python
file = 'most_popular_websites.csv'
with open(file,'w',encoding = 'utf-8') as f:
    headings = ','.join(columns)
    headings += '\n'
    f.write(headings)
    
    for data in rowData:
        rows = ','.join(data)
        rows += '\n'
        f.write(rows)
```

## - Reading the created .csv file using pandas.

```python
data = pd.read_csv(file)
data.head(50)
```
